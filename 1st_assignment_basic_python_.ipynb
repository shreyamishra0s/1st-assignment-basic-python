**Write a python program to convert user input from string to integer,flots and boolean types.

```python
def convert_user_input():
    """
    Prompts the user for input and attempts to convert it to integer, float,
    and boolean types, displaying the results and handling conversion errors.
    """
    # Get string input
    string_input = input("Enter a string: ")
    print(f"Original string: '{string_input}' (Type: {type(string_input)})")

    # Convert to integer
    try:
        int_value = int(string_input)
        print(f"Converted to integer: {int_value} (Type: {type(int_value)})")
    except ValueError:
        print(f"Could not convert '{string_input}' to an integer.")

    # Convert to float
    try:
        float_value = float(string_input)
        print(f"Converted to float: {float_value} (Type: {type(float_value)})")
    except ValueError:
        print(f"Could not convert '{string_input}' to a float.")

    # Convert to boolean
    # For boolean conversion, we'll check for specific string values
    # 'true' (case-insensitive) will be True, and 'false' (case-insensitive) will be False.
    # Other non-empty strings will also evaluate to True when passed to bool().
    # An empty string will evaluate to False.
    normalized_string = string_input.strip().lower()
    if normalized_string == 'true':
        bool_value = True
    elif normalized_string == 'false':
        bool_value = False
    else:
        # Fallback for other non-empty strings
        bool_value = bool(string_input)
    print(f"Converted to boolean: {bool_value} (Type: {type(bool_value)})")

# Run the conversion function
convert_user_input()
```

**Write a program that checks if a number is positive,negative,or zero.

```python
def check_number_status(number):
  """
  Checks if a number is positive, negative, or zero.

  Args:
    number: The number to be checked.

  Returns:
    A string indicating whether the number is "Positive", "Negative", or "Zero".
  """
  if number > 0:
    return "Positive"
  elif number < 0:
    return "Negative"
  else:
    return "Zero"

# Example usage:
num1 = 5
print(f"The number {num1} is: {check_number_status(num1)}")

num2 = -10
print(f"The number {num2} is: {check_number_status(num2)}")

num3 = 0
print(f"The number {num3} is: {check_number_status(num3)}")

num4 = float(input("Enter a number: "))
print(f"The number {num4} is: {check_number_status(num4)}")
```

**Write a pyton program to find the sum of all even numbers between 1 and 50.

```python
def sum_of_even_numbers(start, end):
    """
    Calculates the sum of all even numbers within a given range (inclusive).

    Args:
        start (int): The starting number of the range.
        end (int): The ending number of the range.

    Returns:
        int: The sum of all even numbers in the specified range.
    """
    total_sum = 0
    for number in range(start, end + 1):
        if number % 2 == 0:
            total_sum += number
    return total_sum

# Call the function to find the sum of even numbers between 1 and 50
result = sum_of_even_numbers(1, 50)
print(f"The sum of all even numbers between 1 and 50 is: {result}")
```

**Write a program to reverse a string using a while loop.

```python
def reverse_string_while_loop(input_string):
    """
    Reverses a given string using a while loop.

    Args:
        input_string: The string to be reversed.

    Returns:
        The reversed string.
    """
    reversed_string = ""
    index = len(input_string) - 1  # Start from the last character

    while index >= 0:
        reversed_string += input_string[index]
        index -= 1  # Move to the previous character

    return reversed_string

# Example usage:
original_str = "hello"
reversed_str = reverse_string_while_loop(original_str)
print(f"Original string: {original_str}")
print(f"Reversed string: {reversed_str}")

original_str_2 = "Python"
reversed_str_2 = reverse_string_while_loop(original_str_2)
print(f"Original string: {original_str_2}")
print(f"Reversed string: {reversed_str_2}")
```

**Write a python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Get input from the user
try:
    number = int(input("Enter a non-negative integer: "))
except ValueError:
    print("Invalid input. Please enter an integer.")
else:
    # Handle special cases for factorial
    if number < 0:
        print("Factorial is not defined for negative numbers.")
    elif number == 0:
        print("The factorial of 0 is 1.")
    else:
        # Calculate factorial using a while loop
        factorial = 1
        i = 1
        while i <= number:
            factorial *= i
            i += 1
        print(f"The factorial of {number} is {factorial}")
```
